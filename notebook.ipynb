```python
import numpy as np
import pandas as pd
import ast
import requests
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
```

```python
movies = movies[['title','overview','genres','keywords', 'vote_average', 'id']]
```

```python
movies.shape
```

```python
movies.isnull().sum()
movies.dropna(inplace=True)
movies.isnull().sum()
```

```python
movies.shape
```

```python
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L 
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies['overview']=movies['overview'].apply(lambda x: str(x).split(' '))
movies['title'] = movies['title'].apply(lambda x: str(x).split(' '))
movies['tags']=movies['title'] +movies['overview']+movies['genres']+movies['keywords']
movies['tags'][0]
```

```python
new = movies.drop(columns=['genres','keywords'])
```

```python
posters=new['id']
```

```python
def poster(id):
    res=requests.get('https://api.themoviedb.org/3/movie/{}?api_key=640fb936d11f172ee4527a59a1f488e2'.format(id))
    data=res.json()
    if 'poster_path' not in data.keys():
         return "https://us.123rf.com/450wm/pavelstasevich/pavelstasevich1811/pavelstasevich181101028/112815904-no-image-available-icon-flat-vector-illustration.jpg?ver=6"       
    elif data['poster_path']!=None:
        return "https://image.tmdb.org/t/p/w500"+data['poster_path']
    else:
        return "https://us.123rf.com/450wm/pavelstasevich/pavelstasevich1811/pavelstasevich181101028/112815904-no-image-available-icon-flat-vector-illustration.jpg?ver=6"       

        
        
  
```

```python
l =[]
for i in posters:
    print(i)
    l.append(poster(i))
```

```python
list(enumerate(l))
```

```python
new['posters']=l
```

```python
new['title']=new['title'].apply(lambda x:" ".join(x))
new['overview']=new['overview'].apply(lambda x:" ".join(x))
```

```python
new.shape
```

```python
type(new['overview'])
```

```python
new['tags']=new['tags'].apply(lambda x:" ".join(x))
new['tags'][0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
words =cv.fit_transform(new['tags']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(words)
similarity[0]
```

```python
index = new[new['title'] == 'Avatar'].index[0]
index
```

```python
similarity[0]
```

```python
distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
distances=distances[1:6]
print(distances[0])
type(distances)
distances
        
```

```python
df = pd.DataFrame(distances, columns =['index', 'similarity'])
for i in df['index']:
    print(new.iloc[i].title)
```

```python
def recommend(movie):
    index = new[new['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    print(movie)
    for i in distances[1:6]:
        print(new.iloc[i[0]].title)
        print(new.iloc[i[0]].overview)
        print(new.iloc[i[0]].vote_average)
```

```python
recommend('Batman Begins')
```
